# 06 - Comparacion de modelos

Comparacion entre Lasso Regression y Decision Tree Regressor.

## GUIA RAPIDA PARA EXPONER LA COMPARACION

Este notebook compara los resultados de Lasso Regression y Decision Tree Regressor usando las mismas metricas.

La conclusion del proyecto sale de aqui: Decision Tree Regressor es mejor en esta ejecucion porque obtiene menor `RMSE` y mayor `R2`.

```python
# Carga las metricas generadas durante el entrenamiento de modelos.
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_DIR = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
METRICS_PATH = PROJECT_DIR / 'reports' / 'metricas_modelos.csv'
metrics_df = pd.read_csv(METRICS_PATH)
metrics_df
```

## Comparacion por RMSE

### Comentario para exposicion: RMSE

`RMSE` mide el error de prediccion en la misma unidad de PM 2.5. Si el RMSE es menor, significa que el modelo se equivoca menos en promedio.

```python
# Grafico para comparar visualmente el RMSE de los modelos.
plt.figure(figsize=(7, 5))
plt.bar(metrics_df['modelo'], metrics_df['rmse'])
plt.title('Comparacion de modelos por RMSE')
plt.ylabel('RMSE')
plt.xticks(rotation=15)
plt.show()
```

## Interpretacion breve

El mejor modelo sera el que tenga menor MAE/RMSE y mayor R2. Para este problema, Decision Tree puede capturar relaciones no lineales entre contaminantes, hora y estacion; Lasso ayuda a interpretar relaciones lineales.

### Conclusion

Para elegir el mejor modelo se revisan dos ideas:

- Menor error: menor `MAE` y menor `RMSE`.
- Mejor explicacion: mayor `R2`.

Con los resultados actuales, Decision Tree Regressor gana porque tiene menor RMSE y mayor R2 que Lasso.
